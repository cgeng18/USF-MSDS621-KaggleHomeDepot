```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.feature_extraction import stop_words

from nltk.stem.porter import PorterStemmer
from autocorrect import spell
from scipy import spatial
import operator
import time
```

```python
glove_file = '../data/glove.6B/glove.6B.300d.txt'
def make_dictionary(file):
    '''
    Initiate the glove model as a dictionary
    input: A String which is a file in the project directory
    returns: A dictionary with item = word : 300 d list
    '''
    vecs = dict()
    with open(file) as f:
        lines = f.readlines()
        for word_and_vec in lines:
            elems = word_and_vec.strip().split(' ')
            word = elems[0]
            vec = np.array(elems[1:], dtype=float)
            vecs[word] = vec
    return vecs
```

```python
glove_dic=make_dictionary(glove_file)
```

```python
stops = list(stop_words.ENGLISH_STOP_WORDS)
```

```python
products = pd.read_csv('../data/product_descriptions.csv')
train = pd.read_csv('../data/train.csv', encoding='ISO-8859-1')
train = train.set_index('product_uid').join(products.set_index('product_uid'))
train = train.reset_index()
```

```python
search_term=train['search_term']
```

```python
def clean_term_list(search_terms:list):
    f = open('output.txt', "w")
    f.close()
    for i in range(len(search_terms)):
        t=''
        search_terms[i]=search_terms[i].lower()
        for term in search_terms[i].split(' '):
            if ( term.isalpha() and len(term)>2 and (term not in stops) and (term in glove_dic)):
                t = t +' '+ term
        search_terms[i]=t

        f = open('output.txt', "a")
        f.write(search_terms[i])
        f.write('\n')
        f.close()
```

```python
cleaned=list(train['clean_terms'])  # without transfering into list, the changes made on cleaned will be made on train!
title=list(train['product_title'])
desc=list(train['product_description'])
```

```python
clean_length=np.zeros(len(cleaned))
for i in range(len(cleaned)):
        if not pd.isnull(cleaned[i]):
            clean_length[i]=len(cleaned[i].split(' '))
            
train['clean_length']=clean_length  
```

```python
title_length=np.zeros(len(title))
for i in range(len(title)):
        title_length[i]=len(title[i].split(' '))
            
train['title_length']=title_length    
```

```python
desc_length=np.zeros(len(desc))
for i in range(len(desc)):
        desc_length[i]=len(desc[i].split(' '))
            
train['desc_length']=desc_length  
```

```python
def clean_term_in_doc(terms:list, title:list):
    count=np.zeros(len(terms))
    for i in range(len(terms)):
        if not pd.isnull(terms[i]): 
            title[i]=title[i].lower()
            #print(terms[i])
            for term in terms[i].split(' '):
                #print(term, title[i].split(' '))
                if term in title[i].split(' '):
                    count[i]+=1
    return count
```

```python
c_title=clean_term_in_doc(cleaned, title)
c_desc=clean_term_in_doc(cleaned, desc)
```

```python
train['clean_terms_in_title']=c_title
train['clean_terms_in_desc']=c_desc
```

```python
# firstly, I split the dictionary into a wordlist and a matrix.

wordlist = []
matrix = []
with open(glove_file) as f:
    lines = f.readlines()
    for word_and_vec in lines:
        wordvec = np.array([float(x) for x in word_and_vec.split()[1:]])    
        matrix.append(wordvec / np.linalg.norm(wordvec))
        wordlist.append(word_and_vec.split()[0])
    matrix = np.array(matrix)
```

```python
# I then obtain the unique words that appear in the search_term.
# we have 4804 unique words in the "clean_terms" colum

cleaned = list(train['clean_terms'])
all_words=[]
for t in cleaned:
    all_words+=t.split(' ')

cleaned_set=list(set(all_words))[1:]
len(cleaned_set) # we have 4804 unique words in the "clean_terms" column.
```

```python
def cos_d(v1, v2):
    return 1 - glove_dic[v1].dot(glove_dic[v2].T) / np.linalg.norm(glove_dic[v1]) / np.linalg.norm(glove_dic[v2])
```

```python
# here I count the cos_distance of each word that is in the cleaned_set.
# the output file looks like (each line): w0, w1, w2, w3, w4,
# i didn't print the distance, just the neighbour words
# this will take couple of minutes.

f = open('glove_neighbour_no_w.txt', "w")
f.close()
for word in cleaned_set:
    dots = matrix.dot(glove_dic[word])
    close_index_vec = np.argsort(dots)
    for i in range(5):
        f = open('glove_neighbour_no_w.txt', "a")
        f.write(wordlist[close_index_vec[-1-i]])
        f.write(',')
        #f.write(str(round(cos_d(word, wordlist[close_index_vec[-1-i]]),4)))
        #f.write(',')
        f.close()
    f = open('glove_neighbour_no_w.txt', "a")
    f.write('\n')
    f.close()  
```

```python
# based on the above output file, I then built a dictionary;
# this dictionary stores each word (as key) with its top 4 neighbour words (as value) 

k_dic=dict()
with open('glove_neighbour_no_w.txt') as f:
    lines = f.readlines()
    for line in lines:
        words = line.strip().split(',')
        k_dic[words[0]]=words[1]+' '+words[2]+' '+words[3]+' '+words[4]
    
```

```python
# terms_neighbour is the list which stores the top 4 neighbours of each searching_terms. 
# for example, if the searching term is: cleaned[0]='w1_w2', 
# then the terms_neighbour[0]='n11_n12_n13_n14_n21_n22_n23_n24'.

terms_neighbour=[]
for i in range(len(cleaned)):
    neighbours=''
    if cleaned[i] != '':
        words=cleaned[i].split(' ')
        for w in words:
            neighbours = neighbours+k_dic[w]+' '
    terms_neighbour.append(neighbours)   
```

```python
train['terms_neighbour']=terms_neighbour
```

```python
c_neighbour_title=clean_term_in_doc(terms_neighbour, title)
c_neighbour_desc=clean_term_in_doc(terms_neighbour, desc)
train['neighbours_in_title']=c_neighbour_title
train['neighbours_in_desc']=c_neighbour_desc
```
